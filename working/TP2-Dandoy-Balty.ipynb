## Question 1

Nous allons dans cette première partie montrer que le champ magnétique dans la surface considérée répond à une équation de diffusion. Pour cela considérons l'équation de Maxwell suivante: 

$$\vec{\nabla} \times \vec{B} = \mu_0 \vec{J}$$

En prenant le rotationel de celle-ci on obtient: 

$$  \vec{\nabla} \times (\vec{\nabla} \times \vec{B}) = \vec{\nabla}\times(\mu_{0} \vec{J})$$

Le double rotationnel peut être exprimé comme suit: 

$$\vec{\nabla} \times (\vec{\nabla} \times \vec{B}) = \vec{\nabla} ( \nabla \cdot \vec{B}) - \Delta \vec{B}$$

On sait de plus de part les équations de Maxwell que la divergence du champ magnétique est nulle, et on peut ainsi annuler le premier terme. Il nous reste donc l'égalité: 

$$-\Delta \vec{B} = \vec{\nabla}\times(\mu_{0} \vec{J})$$

Utilisons la relation reliant le courant $\vec{J}$ et le champ électrique $\vec{E}$ ainsi que l'équation de Maxwell nous donnant le rotationnel de $\vec{E}$:

$$-\Delta \vec{B} = \mu_{0} \sigma (\vec{\nabla} \times \vec{E})$$


$$\Delta \vec{B} = \mu_{0} \sigma (\partial_{t} \vec{B})$$

On reconnait bien ici un équation de diffusion vectorielle. Celle-ci peut être grandement simplifiée si on analyse plus en détails la géométrie de notre problème.
On sait dans un premier temps que la longueur infinie du parallélépipède rectangle engendre une symétrie selon l'axe Z (pas de dépendance en z). On pourra donc considérer dans la suite du problème uniquement une section rectangulaire de la structure totale. De plus, on a dans ce cas que le courant circule le long des bords de ce rectangle (dans le sens horlogique), par la loi d'Ampère, on voit donc directement que le problème sera similaire à celui d'un solénoïde et qu'ainsi le champ magnétique sera dirigé uniquement selon l'axe Z (négatif selon les conventions d'axes utilisées dans l'intitulé). Il est donc temps de simplifier notre équation de diffusion en supprimant les composantes selon $x$ et $y$ de notre champ:

$$\frac{\partial^2 B_{z}}{\partial x^2} +\frac{\partial^2 B_{z}}{\partial y^2} = \mu_{0} \sigma (\partial_{t} {B_{z}})$$

On avancera dans la suite du problème, afin d'alléger les notations, en dénotant $B_{z}=B$.



# Question 2

Maintenant que nous disposons de notre équation de diffusion, il nous faut trouver des conditions aux limites du domaine afin de pouvoir complétement la résoudre. Commençons donc par les déterminer sur les bords extérieurs de notre carré explicité plus haut: 

Avant de commencer ceci, rappelons l'équation de discontinuité de la composante tangentielle du champ magnétique entre les deux interfaces que sont l'intérieur du domaine et l'environement extérieur (vue en cours d'électromagnétisme):

$$\vec{n} \times (\vec{B_{ext}} -\vec{B_{int}}) = \mu_{0}\vec{J} $$

où $J$ est la densité de courant de surface ( circulant comme dis plus haut dans le sens horlogique) et vaut $1\frac{A}{m^2}$ dans notre cas.

On sait de plus que le champ magnétique $B_z$ extérieur est nul dans notre configuration. On peut ainsi appliquer sur chacun des bords du carré l'équation de discontinuité de telle sorte que l'on obtient respectivement (en partant du bord en $y=0$ et tournant ensuite dans le sens horlogique):


 $$\vec{u_y} \times \vec{B_{int}} = -\mu_{0}\vec{u_x} \rightarrow B_{z,int} = -\mu_0$$ 

 $$\vec{u_x} \times \vec{B_{int}} = \mu_{0}\vec{u_y}\rightarrow B_{z,int} = -\mu_0$$ 
 
 $$-\vec{u_y} \times \vec{B_{int}} = \mu_{0}\vec{u_x}\rightarrow B_{z,int} = -\mu_0$$ 

 $$-\vec{u_x} \times \vec{B_{int}} = -\mu_{0}\vec{u_y}\rightarrow B_{z,int} = -\mu_0$$ 

On trouve donc systématiquement la condition que le champ magnétique dois être orienté selon l'axe $z$ (négatif selon le choix des axes) et de module $\mu_0$. Il s'agit donc de conditions de Dirichlet aux bords extérieurs du domaine. 


Intéressons nous à présent aux conditions sur le supraconducteur: 

Dans un premier temps, intéressons nous à la valeur du champ magnétique en tout temps au sein du supraconducteur. On sait que que le courant de surface est relié au champ magnétique selon l'équation:

$$\vec{J} = \sigma \vec{E}$$

or dans le supraconducteur, $\sigma = \infty$ ce qui engendre que le champ électrique doit être nul afin d'êmpecher toute divergence du courant. Ainsi $\vec{E}$ est nul en tout temps. De plus, on sait que le champ électrique est relié au champ magnétique par:

$$\vec{\nabla} \times \vec{E} = -\partial_t \vec{B}$$

Ainsi, si $\vec{E}$ est nul en tout temps, le champ magnétique demeurera constant dans le supraconducteur. Puisque $\vec{B}$ est nul initialement, il le restera en tout temps (dans le supraconducteur).

Mais que se passe t'il sur le bord du conducteur, à la frontiere avec le supraconducteur? Pour le savoir, on utilise à nouveau un équation de continuité, mais pour le champ électrique cette fois-ci. On sait que le champ est continu dans sa composante tangentielle, et donc nous avons l'équation: 

$$\vec{n} \times (\vec{E_{int}} -\vec{E_{supra}}) =0 $$

Par ce qu'on a démontré plus haut, cette expression se simplifie et on obtient: 


$$\vec{n} \times (\vec{E_{int}})=0 $$


$$\vec{n} \times (\vec{J})=0 $$

On obtient donc dans un premier temps une contrainte sur $\vec{J}$. En effet, cette condition de continuité sur le champ électrique exprimée plus haut, nous renseigne sur les valeurs du champ à des distances très proches du bord des deux côtés de l'interface (en réalité nous faisons tendre ces distances vers 0). Dans le cas présent, nous en déduisons que la valeur tangentielle du courant est nulle sur le bord, côté conducteur. Plus précisement: 

$$J_x = 0$$ 

$$J_y=0 $$

Respectivement le long des horizontaux et verticaux.

Or on sait également que $\vec{\nabla} \times \vec{B} = \mu_0 \vec{J}$ et donc:


$$\vec{n} \times (\vec{\nabla} \times \vec{B})=0 $$

Ainsi on peut affirmer que la valeur tangentielle du rotationnel doit être nulle sur le bord du conducteur, de telle sorte que l'on obtient des conditions de Neuman du type (en considérant le fait que $\vec{B} = B \vec{u_z}$):

$$\frac{\partial B}{\partial y} =0$$ 

$$\frac{\partial B}{\partial x} =0$$ 

Respectivement le long des axes horizontaux et verticaux.

Nous disposons à présent de tous les ingrédients nous permettant de résoudre numériquement cette équation de diffusion. 


## Question 3

Commençons par importer toutes les fonctions dont nous aurons besoin.

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams, cm
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
from matplotlib import animation
from IPython.display import HTML
```

Pour résoudre le problème, nous allons appliquer la méthode "forward-time, centered scheme", avec une discrétisation centrée d'ordre 2, vue dans le cours (nous justifions légerement plus bas que nous pouvons utiliser cette méthode dans risque d'instabilités). Seulement, nous ajouterons une petite subtilité: puisqu'à la question 4 nous devons faire une animation de l'évolution du champ magnétique, nous allons sauver à intervalle régulier des solutions. Pour ce faire, ajoutons un argument à notre fonction. Cet argument sera un tableau de tableau que nous remplirons au fur et à mesure avec le champ magnétique calculé toutes les 100 itérations (B intermédiaire--> Bint). On arrêtera de stocker si tout le tableau a été rempli.

Voyons à présent comment implémenter nos conditions explicitées à la section précedente:

° Tout d'abord, on a prouvé que $\vec{B}$ doit être nul en tout temps dans l'intégralité du supraconducteur. Cela se traduit par le fait que l'on doit imposer dans toute la surface du supraconducteur (y compris le bord) un champ magnétique nul.

° On a développé plus haut l'existence de conditions de Neumann sur le bord, côté conducteur. Or au point ci-dessus, on vient d'imposer un champ magnétique nul en tout temps sur le bord, ce qui revient à des conditions de Dirichlet et donc a une incompatibilité avec celles de Neumann. D'où provient l'incohérence? 

En réalité, il n'y en a tout simplement pas. En effet, comme expliqué plus haut, les conditions de continuité à l'interface entre deux milieux, nous renseignent sur les valeurs du champ à des distances infinitésimales de part et d'autre de l'interface. Or dans notre cas, nous travaillons dans un modèle où notre élément infinitésimal n'est pas infiniment petit... On ne pourra donc pas considérer que la condition opère exactement sur le bord mais bien à une distance $dx$ du bord , où ce $dx$ représente justement notre élément de discrétisation. Ainsi, tout se passe comme si on imposait notre condition de Neumann non pas sur le bord mais sur le contour à une distance $dx$ du bord.

° Enfin, en ce qui concerne la condition de Dirichlet sur le bord extérieur du conducteur, on peut imposer la valeur du champ magnétique en tout temps en dehors de la fonction qui suit. 

```python
def ftcs_save(T,nt,dt,dx,dy,alpha,Bint):
    
    for n in range (nt):
        
        Tn=T.copy()
        nbIm=len(Bint)
        
        if (n/100)==int(n/100) and n/100<nbIm:    # Si la n-eme itération divisé par 100 est un entier inférieur
            Bint[int(n/100)]=Tn                   # à la taille de Bint, nous stockons la valeur du champ 
                                                  # magnétique en cette itération.
        
        
             
        T[1:-1,1:-1] = Tn[1:-1,1:-1] + alpha *\
            (dt/dy**2 * (Tn[2:,1:-1] - 2*Tn[1:-1,1:-1] + Tn[:-2,1:-1]) +\
             dt/dx**2 * (Tn[1:-1,2:] - 2*Tn[1:-1,1:-1] + Tn[1:-1,:-2]))
            
            
        T[40:-40,40:-40]=0                       # Champ magnétique nul dans le supraconducteur en tout temps.
        T[38:-38,38:-38][1,:]=T[38:-38,38:-38][0,:]  # Implémentation des conditions de Neumann sur le bord intérieur.
        T[38:-38,38:-38][-2,:]=T[38:-38,38:-38][-1,:]
        T[38:-38,38:-38][:,1]=T[38:-38,38:-38][:,0]
        T[38:-38,38:-38][:,-2]=T[38:-38,38:-38][:,-1]
        
               
    return T
```

Nous imposons à présent nos conditions initiales dont celles de Dirichlet aux bords exterieurs

```python
# Conditions initiales

L = 1.5e-2             # Longueur du carré


nx = 121               # On discrétise x en 121 points
ny = 121               # On discrétise y en 121 points
nt = 50000             # Pour l'exercice nous devrons connaitre la valeur de B après 50 000 itérations

dx = L/(nx-1)         # Pas de x
dy = L/(ny-1)         # Pas de y
dt = 2.5*10**(-7)     # Pas de temps
    
x = numpy.linspace(0,L,nx)
y = numpy.linspace(0,L,ny)

m=1.26*10**(-6)        # mu_0
s=5.96*10**7           # sigma
alpha = 1/(s*m)


Bi = numpy.zeros((ny, nx))
Bi[0,:]=-m                       # Conditions de Dirichlet aux bords externes
Bi[:,0]=-m
Bi[-1,:]=-m
Bi[:,-1]=-m

Bint= numpy.zeros((51,nx,ny))        # Tableau avec les valeurs intermédiaires de B
```

Le pas de temps choisi est très petit pour avoir une stabilité de la solution. Pour rappel, afin d'éviter que la solution explose, il faut satisfaire à la condition: $\frac{\alpha \Delta t}{\Delta x^2} = \frac{\alpha \Delta t}{\Delta y^2} < \frac{1}{4}$. Avec nos valeurs de $dt$ et $dx(=dy)$, nous obtenons: $\frac{\alpha \Delta t}{\Delta x^2} = \frac{\alpha \Delta t}{\Delta y^2} = 0.2130$ donc notre pas de temps est justifié.

On suppose qu'après $1,25\cdot10^{-3}$ secondes, nous aurons déjà une belle animation, d'où l'assignation de 51 tableaux dans Bint.


Calculons:

```python
B = ftcs_save(Bi.copy(),nt,dt,dx,dy,alpha,Bint)         # Calcul
```

Une fois que nous avons tout calculé nous aimerions avoir un aperçu du système après évolution.

```python
# Champ B après 1.25 x 10^-2 s

pyplot.figure(figsize=(8,7))
pyplot.contourf(x,y,B,1000,cmap=cm.viridis)
pyplot.xlabel('$x [m]$')
pyplot.ylabel('$y [m]$')
pyplot.title('Champ B.1z dans le plan (x,y)' , fontsize=18)
pyplot.colorbar();
```

La barre colorée est exprimée en Tesla. 

Essayons de comprendre pourquoi un léger rectangle bleu clair est présent entre la surface mauve et la surface jaune. Comme expliqué plus haut, nous appliquons nos conditions de Neumann non pas exactement sur le bord mais sur une rectangle légerement plus grand. Mais que se passe-t-il dans la region entre ce rectangle et le bord du supraconducteur? On peut imaginer que le champ évolue perpetuellement comme une moyenne entre les deux bords, d'où la coloration turquoise.

Nous allons dans cette partie tenter de calculer le courant induit dans la surface du conducteur. Pour cela rappelons l'équation de Maxwell nous donnant le lien entre le champ magnétique local et la densité de courant de surface qui s'y trouve:

$$\vec{\nabla} \times \vec{B} = \mu_0 \vec{J}$$

Développons donc ce rotationnel en sachant que $\vec{B}$ a uniquement une dépendence en $z$ : 

$$ \frac{\partial B}{\partial y} \vec{u_x} - \frac{\partial B}{\partial x} \vec{u_y} = \mu_0 J_x \vec{u_x} + \mu_0 J_y \vec{u_y}$$

$\vec{J}$ n'a donc pas de dépendence en $z$. On va ainsi résoudre notre problème en essayant de trouver les composantes en $x$ et en $y$ de notre courant.

Pour cela, et puisque nous disposons d'une tableau des différentes valeurs du champ magnétique à travers la surface, nous allons utiliser notre discrétisation pour résoudre ces équations, de telle sorte que: 


$$\frac{ B^{x, y+1} - B^{x,y}}{\Delta y} = \mu_0 J^{x,y}_x $$


$$\frac{ B^{x+1, y} - B^{x,y}}{\Delta x} = \mu_0 J^{x,y}_y $$

Où l'exposant représente la position en $x,y$ tandis que l'indice représente la composante du vecteur $J$. En utilisant directement la discrétisation utilisée dans la résolution de l'équation de diffusion du champ magnétique, on trouve que le courant en ($2,5\cdot10^{-3} m;2,5\cdot10^{-3} m;0m$) après $1,25.10^{-3}s$ est directement lié au champ magnétique au sein de $Bint[50]$. ## Question 7

Ici nous cherchons la valeur du flux magnétique après $1,25\cdot 10^{-2}s$. Rappelons que le flux magnétique traversant une surface se calcule en prenant le produit scalaire entre le champ magnétique et la surface (avec comme vecteur directeur le vecteur normal à la surface). Dans notre cas, le champ magnétique et le vecteur normal à la surface sont parallèles, ainsi le produit scalaire est simplement un produit des normes.

Nous possédons un carré découpé en 120x120 plus petits carrés et connaissons la valeur du champ magnétique pour chacun des sommets des carrés. Nous allons supposer que le champ magnétique est constant sur ces carrés et prendrons comme valeur de B la moyenne arithmétique des valeurs aux 4 sommets. Nous allons calculer le flux magnétique pour chacun des carrés et ensuite tout sommer. Remarquons que, si on met en évidence l'élément de surface, on peut d'abord tout sommer et ensuite faire le produit.

```python
Btot=0        # sera la somme de tous les champ magnétique sur les carrés
ds=dx*dy      # élément de surface

for i in range (nx-1):           # On regarde sur les abscisses
    for j in range(ny-1):        # et sur les ordonnées
        
        Btot=Btot+1/4*(B[i,j]+B[i+1,j]+B[i,j+1]+B[i+1,j+1])     # Somme des moyennes arithmétiques du champ magnétique
                                                                # sur chaque sommet du carré considéré
        
        
print( 'Le flux magnétique après 1,25.e-2s vaut [T.m^2]:')
Btot*ds
```

La norme de valeur est très petite. Ce n'est pas étonnant! Tout d'abord, il n'est pas simple d'obtenir un champ magnétique élevé. Au CMS par exemple, les valeurs atteintes sont de l'ordre de 4 Tesla et c'est déjà beaucoup (en question 5 nous avions obtenu un champ magnétique de l'ordre $10^{−6}$T). De plus, la surface considérée n'est pas grande non plus: un carré de 1,5 cm de nous donne une aire de l'ordre de $10^{-4}m^2$.
La valeur obtenue est alors tout à fait cohérente.

Nous allons dans cette partie tenter d'expliciter une autre méthode de détermination du flux magnétique. Nous allons également tenter de déterminer l'allure de son évolution temporelle au sein de la surface.

Contrairement à la méthode utilisée ci-dessus, nous allons plutôt sommer les valeurs du champ magnétique en tous les points contenus dans notre tableau B calculé plus haut (valeurs du champ magnétique à 0,0125 s) et le multiplier par un élément de surface $ds$ qui ne sera rien d'autre que notre $dx \cdot dy$ (chaque point se situe en réalité au centre de son élément de surface). Cependant en effectuant ceci, on introduit une erreur avec les points de notre tableau localisés sur le bord. En effet ceux-ci voient leur élément de surface associé sortir de moitié (pour les points le long du bord) ou d'un quart (pour ceux situé sur les coins). Une correction serait alors de supprimer ces morceaux de surfaces excedentaires.

Voyons comment implémenter ceci:

```python
def Flux(A):
    
    X=0
    ds=dx*dy
    for i in range (nx):
        for j in range (ny):
            X=X+A[j,i]
    Z=4*119*(1/2)*(-m) + 4*(3/4)*(-m)           #correction due à la surface au bord qu'on ne prend pas 
    B=X-Z                                       #on enlève la valeur du champ B pour les morceaux qui sortent 
    return B*ds
```

```python
print ('Le flux du champ magnétique après 1,25.e-2s est donné par [Tm2]:')
Flux(B)
```

Remarquons que nos solutions sont très similaires.
